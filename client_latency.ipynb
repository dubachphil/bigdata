```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("latency_p_min.csv",
                   sep='\t',
                   names=["time","avg_lat"],
                  header=None)
```

```python
data.head()
```

```python
data["avg_lat"].max()
```

```python
data["avg_lat"].min()
```

```python
plt.figure(figsize=(15,10))
plt.hist(data["avg_lat"],alpha=0.5,bins=30)
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.distplot(data["avg_lat"], hist=False)
```

```python
# Einbeziehen der Population mit Farben
data.plot(kind="scatter",x="time",y="avg_lat", alpha=0.3,
            s=30, label="avg_lat", figsize=(15,10),
            c=data["avg_lat"], cmap=plt.get_cmap("jet"), colorbar=True)
plt.legend()
```
