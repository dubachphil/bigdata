```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/dubachphil/bigdata/master/latency_p_min.csv",
                   sep='\t',
                   names=["time","avg_lat"],
                  header=None)
```

```python
data.head()
```

```python
data["avg_lat"].max()
```

```python
data["avg_lat"].min()
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.distplot(data["avg_lat"], hist=True, kde = True, kde_kws = {'shade': True, 'linewidth': 1}, label = "Anteil")
sns.distplot(data["avg_lat"].quantile([.25, .5, .75]), hist=False, kde = True, kde_kws = {'shade': True, 'linewidth': 1}, label = "Quantille 25/50/75")

plt.legend(prop={'size': 16}, title = 'Legende')
plt.title('Density Plot - Anteil Latenz pro Minute')
plt.xlabel('Durchschnitt Latenz pro Minute')
plt.ylabel('Density')
```

```python
plt.figure(figsize=(15,10))
plt.hist(data["avg_lat"],alpha=0.5,bins=15)
```

```python
# Einbeziehen der Population mit Farben
data.plot(kind="scatter",x="time",y="avg_lat", alpha=1,
            s=30, label="Durchschnitt Latenz pro Minute in ms", figsize=(15,10),
            c=data["avg_lat"], cmap=plt.get_cmap("jet"), colorbar=True)
plt.legend()
```
